```python
import numpy as np
import keras
import keras_contrib
from numpy import loadtxt
from keras.models import load_model
import time
 
# load model
model_path='results\\211219_24jam\\encoder\\TSC_itr_1\\F2MD2NPY\\best_model.hdf5'
model = keras.models.load_model(model_path,custom_objects={'InstanceNormalization':keras_contrib.layers.InstanceNormalization})
# summarize model.
model.summary()
# load dataset
X = np.load('saveFile_D40_V2\\valuesSave24.npy')
#Y = np.load('saveFile_D40_V2\\targetSave24.npy')
# evaluate the model
X = X.reshape((X.shape[0], X.shape[1], 1))
#score = model.evaluate(X, Y, verbose=0)
start_time = time.time()
y_pred = model.predict(X)
end_time = time.time()-start_time
print(y_pred)
print(end_time)
```
